```python
import sys
    
class Budget():
    def __init__(self,amount,category,balance):
        self.amount=amount
        self.category=category.title()
        self.balance=balance
    
    def expense(self):
        print(f"You made a purchase:")
        nbal=self.balance-self.amount
        print(f"You spent {self.amount} on {self.category}. Your updated balance is {nbal}.")
        return nbal
    
    def deposit(self):
        print(f"You made a deposit:")
        nbal=self.balance+self.amount
        print(f"You deposited {self.amount} for {self.category}. Your updated balance is {nbal}.")
        return nbal
    
    def checkbalance(balance):
        print(f"Your current balance is {balance}.")
        return balance
    
counter=0
while True:
    start=input("Would you like to use the Expense Tracker? Enter y/n. ")
    if start=='y':
        counter=counter+1
        if counter==1:
            amount=int(input("Enter the amount: "))
            category=input("Enter the category: ")
            balance=int(input("Enter your starting balance: "))
            entry1=Budget(amount,category,balance)
            entrytype=input("Would you like to track an Expense (press E), a Deposit (press D), or Check your Balance (press C)?")
            if entrytype=='e':
                entry=Budget(amount,category,balance)
                balance=entry.expense()
            elif entrytype=='d':
                entry=Budget(amount,category,balance)
                balance=entry.deposit()
            elif entrytype=='c':
                entry=Budget(amount,category,balance)
                balance=entry.checkbalance()
        elif counter>=2:
            entrytype=input("Would you like to track an Expense (press E), a Deposit (press D), or Check your Balance (press C)?")
            if entrytype=='e':
                amount=int(input("Enter the amount: "))
                category=input("Enter the category: ")
                newentry=Budget(amount,category,balance)
                balance=newentry.expense()
            elif entrytype=='D'or'd':
                amount=int(input("Enter the amount: "))
                category=input("Enter the category: ")
                newentry=Budget(amount,category,balance)
                balance=newentry.deposit()
            elif entrytype=='c':
                entry=Budget(amount,category,balance)
                balance=entry.checkbalance()

    elif start=='n':
        break
```
